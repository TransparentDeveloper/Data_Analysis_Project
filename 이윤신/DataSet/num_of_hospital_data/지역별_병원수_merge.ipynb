# date 삽입하기

```python
pretreatment2 = pd.DataFrame()
```

```python
def yyyymmdd_insert(tot_date, i):
  global pretreatment2
  for j in range(1,tot_date+1):
      date = ''
      if j<10:
        date = '0'+str(j)
      else:
        date = str(j)
      yyyymmdd = str(pretreatment.iloc[i,7]) + date
      new_data_frame = pd.DataFrame({
          'area':pretreatment.iloc[i:i+1,0],
          'general_hospits_num':pretreatment.iloc[i:i+1,1],
          'general_hospits_bed':pretreatment.iloc[i:i+1,2],
          'hospits_num':pretreatment.iloc[i:i+1,3],
          'hospits_bed':pretreatment.iloc[i:i+1,4],
          'affiliated_hospits_num':pretreatment.iloc[i:i+1,5],
          'affiliated_hospits_bed':pretreatment.iloc[i:i+1,6],
          'yyyymmdd': yyyymmdd
      })
      pretreatment2 = pd.concat([pretreatment2, new_data_frame])
  

for i in range(0,len(pretreatment)):
  month = int(pretreatment.iloc[i,7]) % 100
  if(month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12):
    yyyymmdd_insert(31,i)
  elif (month == 4 or month == 6 or month == 9 or month == 11):
    yyyymmdd_insert(30,i)
  else:
    year = int(int(pretreatment.iloc[i,7])/100)
    if(year == 2012 or year == 2016):
      yyyymmdd_insert(29,i)
    else:
      yyyymmdd_insert(28,i)
```
